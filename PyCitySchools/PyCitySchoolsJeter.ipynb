```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

```python
# Rename column titles 

renamed_school_data_complete_df = school_data_complete.rename(columns = {"student_name": "Student Name", "school_name": "School Name",
                                                                        "reading_score": "Reading Score", "math_score": "Math Score",
                                                                        "gender": "Gender", "grade":"Grade", "type":"School Type", 
                                                                         "size": "# of Students", "budget":"Budget"})

#renamed_school_data_complete_df["Overall Score"] = (renamed_school_data_complete_df["Reading Score"]+renamed_school_data_complete_df["Math Score"])/2
#renamed_school_data_complete_df.head()
renamed_school_data_complete_df.head()
```

```python
# Calculate the total number of schools, number of students, and total budget

school_count = len(school_data["school_name"])
student_count = len(student_data["Student ID"])
budget_total = school_data["budget"].sum()
format_budget_total = "${:,.2f}".format(budget_total)
#print(f"Number of Schools: {school_count}")
#print(f"Number of Students: {student_count}")
#print(f"Total Budget: {format_budget_total}")

# Calculate the average math score and average reading score
average_math_score = round(renamed_school_data_complete_df["Math Score"].mean(), 1)
average_reading_score = round(renamed_school_data_complete_df["Reading Score"].mean(), 1)
#print(f"Average Math Score: {average_math_score}")
#print(f"Average Reading Score: {average_reading_score}")

# Calculate the percentage of students with a passing math score and a passing reading score (70 or greater).
# David Jenkins (TA) helped with syntax for count of passing students

pass_read_count = renamed_school_data_complete_df.loc[renamed_school_data_complete_df["Reading Score"] >= 70]['Reading Score'].count()   
pass_math_count = renamed_school_data_complete_df.loc[renamed_school_data_complete_df["Math Score"]>=70]['Math Score'].count()  
pass_overall_count = renamed_school_data_complete_df.loc[(renamed_school_data_complete_df["Math Score"]>=70) & (renamed_school_data_complete_df["Reading Score"]>= 70)]["Student Name"].count()  
#print(f"Count passing reading: {pass_read_count}"
#print(f"Count passing math: {pass_math_count}")

pass_math = pass_math_count/student_count
percent_pass_math = "{:.0%}".format(pass_math)
pass_read = pass_read_count/student_count
percent_pass_read = "{:.0%}".format(pass_read)
#print(percent_pass_math)
#print(percent_pass_read)

      
# Calculate the percentage of students who passed math and reading (% Overall Passing)
# Insert Column of "Overall Score" which is the average score of the Reading & Math scores

print(pass_overall_count)

 
# Calculate count of students passing Overall

print(pass_overall_count)
pass_overall = pass_overall_count/student_count
percent_pass_overall = "{:.0%}".format(pass_overall)
print(percent_pass_overall)
```

```python
#Create a dataframe to hold the above results  Optional: give the displayed data cleaner formatting
distract_summary_df =pd.DataFrame({
    "Total School": [school_count],
    "Total Students": [student_count],
    "Total Budget": [format_budget_total],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_pass_math],
    "% Passing Reading": [percent_pass_read],
    "% Passing Overall": [percent_pass_overall]
}) 
distract_summary_df.head()

# Note: Check Overall percentage math.
```

```python

# renamed_school_data_complete_df
grouped_school_df = renamed_school_data_complete_df.groupby(['School Name'])

school_avg_reading = round(grouped_school_df["Reading Score"].mean(), 1)
school_avg_math = grouped_school_df["Math Score"].mean()
school_count_students = grouped_school_df["Student ID"].count()

# Budget
school_budget = grouped_school_df["Budget"].max()

# needed to calc percentages - Andrew Perez helped me with this syntax
#school_pass_math_count = renamed_school_data_complete_df[renamed_school_data_complete_df["Math Score"] >= 70].groupby("School Name")
school_pass_math_count = renamed_school_data_complete_df.loc[renamed_school_data_complete_df["Math Score"]>69].groupby("School Name")
school_pass_reading_count = renamed_school_data_complete_df[renamed_school_data_complete_df["Reading Score"] >= 70].groupby("School Name")
school_pass_overall_count = renamed_school_data_complete_df[(renamed_school_data_complete_df["Reading Score"] >= 70) & (renamed_school_data_complete_df["Math Score"] >= 70)].groupby("School Name")


#Calculate the percentage of students with a passing math, reading & overall score (70 or greater) - Doris Meirs helped me find the sytax error by sharing her code.
school_pass_math_percent= school_pass_math_count['Student ID'].count()*100/school_count_students
school_pass_reading_percent = school_pass_reading_count['Student ID'].count()*100/school_count_students
school_pass_overall_percent = school_pass_overall_count['Student ID'].count()*100/school_count_students

# Toformat output on table later
#school_pass_math_percent = grouped_school_df["% Passing Math"].count()
#format_school_budget ="${:,.2f}".format(school_budget)
 
#generate data frame

grouped_school_df = pd.DataFrame({"Number of Students": school_count_students,
                                  "Average Reading Scores": school_avg_reading,
                                  "Average Math Scores": school_avg_math,
                                  "Percent Passing Reading": school_pass_reading_percent,
                                  "Percent Passing Math": school_pass_math_percent,
                                  "Percent Passing Overall": school_pass_overall_percent,
                                  "Budget": school_budget                                  
                                 })

grouped_school_df.head()
```
